```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"accidents_2005_to_2007.csv.zip"
data = pd.read_csv(file_path, compression='zip')
```

```python
data
```

```python
selected_columns = ['Weather_Conditions', 'Road_Surface_Conditions', 'Special_Conditions_at_Site', 'Urban_or_Rural_Area', 'Time']
selected_data = data[selected_columns].copy()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Weather_Conditions', data=selected_data)
plt.title('Accident Counts by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Road_Surface_Conditions', data=selected_data)
plt.title('Accident Counts by Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Time'] = pd.to_datetime(data['Time'])
data['Hour'] = data['Time'].dt.hour
plt.figure(figsize=(10, 6))
sns.countplot(x='Hour', data=data)
plt.title('Accident Counts by Time of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Accident Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Special_Conditions_at_Site', data=selected_data)
plt.title('Accident Counts by Special Conditions at Site')
plt.xlabel('Special Conditions at Site')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Urban_or_Rural_Area', data=selected_data)
plt.title('Accident Counts by Urban or Rural Area')
plt.xlabel('Urban or Rural Area')
plt.ylabel('Accident Count')
plt.show()
```
